```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#Load the data set
df = pd.read_csv("E:\\Data sciences notes\\Assignment\\Logistic Regression\\bank-full.csv",sep=';')
df.head()
```

```python
data=pd.get_dummies(df,columns=['job','marital','education','contact','poutcome'])
data
```

```python
data.info()
```

```python
# Custom Binary Encoding of Binary o/p variables 
data['default'] = np.where(data['default'].str.contains("yes"), 1, 0)
data['housing'] = np.where(data['housing'].str.contains("yes"), 1, 0)
data['loan'] = np.where(data['loan'].str.contains("yes"), 1, 0)
data['y'] = np.where(data['y'].str.contains("yes"), 1, 0)
data
```

```python
#Replace Encoding for month categorical varaible
data['month'].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
data=data.replace(order)
data
```

```python
data.info()
```

```python
# Dividing our data into input and output variables
x=pd.concat([data.iloc[:,0:11],data.iloc[:,12:]],axis=1)
y=data.iloc[:,11]
```

```python
# Logistic regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

model predictions

```python
# Predict for x dataset
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

Testing model accuracy

```python
# Confusion Matrix for the model accuracy
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39093+1274)/(39093+829+4015+1274)
```

```python
# As accuracy = 0.8928, which is greater than 0.5
classifier.predict_proba(x)[:,1]
```

```python
# ROC Curve plotting and finding AUC value
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
